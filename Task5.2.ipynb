```python
import matplotlib.pyplot as plt
```

```python
labels = ['Android', 'iOS', 'Others']
colors=['#4da6ff', '#1f78b4', '#0059b3']
shares = [71, 27, 2]
```

```python
plt.pie(shares,labels=labels,autopct='%1.1f%%',colors=colors,wedgeprops={'edgecolor': 'black', 'linewidth': 2},startangle=140)
plt.title('Mobile OS Market Share')
plt.tight_layout()
plt.show()
```

```python
explode = [0, 0.1, 0]

plt.pie(shares,labels=labels,autopct='%1.1f%%',colors=colors,wedgeprops={'edgecolor': 'black', 'linewidth': 2},startangle=140,explode=explode)

plt.title("Exploded Pie Example (iOS)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.barh(labels, shares, color=['#4da6ff', '#1f78b4', '#0059b3'])
plt.xlabel('Market Share')
plt.title('Mobile OS Market Share')
plt.tight_layout()
plt.show()
```

### After comparing both charts, I think the horizontal bar chart is easier to understand especially when you want to compare values and have more than three categories.The pie chart is visually appealing, but it is less precise for comparison.But in this case, since we only have three items to visualize, both charts are appropriate and easy to read.
