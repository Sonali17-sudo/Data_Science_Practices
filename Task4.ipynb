```python
import matplotlib.pyplot as plt
```

# Data for plotting

```python
students = ['Sonali', 'Mansi', 'Amruta', 'Raksha', 'Ratna']
marks = [99, 97, 85, 92, 88]
plt.plot(students,marks)
plt.show()
```

# Adding X and Y Labels

```python
plt.plot(students,marks)
plt.xlabel("Student Names")
plt.ylabel("Marks Scored")
plt.title("Result of the students")
plt.show()
```

# Multiple Lines in One Plot

```python
students = ["Test 1", "Test 2", "Test 3", "Test 4"]
Sonali = [99, 98, 97, 99]
Mansi = [87, 92, 89, 90]
plt.plot(students, Sonali, label="Sonali")
plt.plot(students, Mansi, label="Mansi")
plt.xlabel("Tests")
plt.ylabel("Marks Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

# Using Format Strings

```python
plt.plot(students, Sonali, 'ro--',label="Sonali")
plt.plot(students, Mansi, 'g^:',label="Mansi")
plt.legend()
plt.show()
```

# Color and Line Style Arguments

```python
Amruta = [95, 90, 93, 92] 
plt.plot(students, Sonali, color='Magenta', linestyle='--', label="Sonali")
plt.plot(students, Mansi, color='green', linestyle='-.', label="Mansi")
plt.plot(students, Amruta, color='blue', label="Amruta")
plt.xlabel("Tests")
plt.ylabel("Marks Scored")
plt.title("Performance Comparison of Students")
plt.legend()
plt.grid(True)
plt.show()
```

# Line Width and Layout Tweaks

```python
plt.plot(students, Sonali, linewidth=2, label="Sonali")
plt.plot(students, Mansi, linewidth=5, label="Mansi")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# XKCD Comic Style

```python
with plt.xkcd():
    plt.plot(students, Sonali, label="Sonali")
    plt.plot(students, Mansi, label="Mansi")
    plt.xlabel("Tests")
    plt.ylabel("Marks Scored")
    plt.title("Performance in XKCD Style")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
